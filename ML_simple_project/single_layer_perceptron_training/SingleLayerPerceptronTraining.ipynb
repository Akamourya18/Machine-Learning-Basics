## Problem Statement



Train a single layer perceptron to predict a mathematical equation 

`y = x^2`

## Generate Dataset 


The generated dataset should have data points as shown, which can easily be created using python

```
x = [0, 1, 2, 3, 4,...]

y = [0, 1, 4, 9, 16,...]
```

````python
x = [i for i in range(1000+1)] #list comprehension
print(x)
````

````python
y = [i*i for i in range(1000+1)]
print(y)
````

## Splitting Dataset into training and testing set

````python
x_train = x[:-250] #Training Data (75%)
y_train = y[:-250] #Training Labels

x_test = x[-250:] #Testing Data (25%)
y_test = y[-250:] #Testing Labels

print(f'''
Training Data:

xTrain : {x_train}
yTrain : {y_train}

Testing Data:

xTest : {x_test}
yTest : {y_test}
''')
````

## Model Training

````python
# Importing libraries

import tensorflow as tf
from tensorflow import keras
````

While defining the model,

- `units` here denotes the number of neurons 

- `input_shape` denotes the shape of the input data that you provide your neuron here (1D Array)

````python
#perceptron model

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1, input_shape=[1]) 
])
````

To compile the model,

- The optimizer used is [adam](https://keras.io/api/optimizers/adam/). There are a tons of other optimizers available but the main aim is to use these methods to change the attributes of your model such as weights and learning rate in order to reduce the losses.

- The loss function used is mean absolue error `mae` since this is a regression problem. 

$${\displaystyle \mathrm {MAE} ={\frac {\sum _{i=1}^{n}\left|y_{i}-x_{i}\right|}{n}}}
$$

where,

- $y_{i}$ denotes the true label (the labels you have)
- $x_{i}$ denotes the predicted labels that the model outputs 


````python
model.compile(optimizer='adam', loss='mae')
````

We define `epochs` i.e. the number of times the training set passes through the learning algorithm.

We fit our model on the training data with the aim of reducing error `val_error`. Error here is the difference between the predicted label (output of the model) and the actual label (labels we have)


````python
epochs = 8000 

model.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), epochs= epochs)
````
